```python
import numpy as np
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras import layers
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.losses import CategoricalCrossentropy
```

```python
(X_train, y_train), (X_test, y_test) = keras.datasets.mnist.load_data()
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
X_train = X_train.astype("float32") / 255
X_test = X_test.astype("float32") / 255
```

```python
y_train = to_categorical(y_train, num_classes=10)
y_test = to_categorical(y_test, num_classes=10)
```

```python
print(y_train.shape)
print(y_test.shape)
```

```python
model = keras.Sequential()
model.add(layers.Input(shape=(28, 28)))
model.add(layers.Flatten())
model.add(layers.Dense(256, activation='relu'))
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(10, activation='softmax'))
model.compile(loss=keras.losses.categorical_crossentropy,
              optimizer=keras.optimizers.Adam(learning_rate=6e-5),
              metrics=['accuracy'])
```

```python
train_portion = 0.05
model.save("model.keras")

history_1 = {'train_loss':[], 'train_acc':[], 'val_loss':[], 'val_acc':[], 'test_loss':[], 'test_acc':[]}
while train_portion <= 1:
  X,  y = X_train[0:int(train_portion * 60000)] , y_train[0:int(train_portion * 60000)]
  model = keras.saving.load_model("model.keras")
  history = model.fit(X, y, epochs=25, batch_size= 10, validation_split=0.1)
  test_loss, test_acc = model.evaluate(X_test, y_test, batch_size=32)
  history_1['train_loss'].append(history.history['loss'][-1])
  history_1['train_acc'].append(history.history['accuracy'][-1])
  history_1['val_loss'].append(history.history['val_loss'][-1])
  history_1['val_acc'].append(history.history['val_accuracy'][-1])
  history_1['test_loss'].append(test_loss)
  history_1['test_acc']  .append(test_acc)
  train_portion += 0.15
model.save('Model.keras')
```

```python
def plot_loss_acc(history):
    train_loss = history['train_loss']
    test_loss = history['test_acc']
    val_loss = history['val_loss']
    val_acc = history['val_acc']
    train_acc = history['train_acc']
    test_acc = history['test_acc']

    epochs = range(1, len(train_loss) + 1)

    fig, axs = plt.subplots(2, figsize=(10, 7))
    fig.suptitle('Training and Validation and Test Metrics')

    # Loss plot
    axs[0].plot(epochs, train_loss, label='Training loss', c='red')
    axs[0].plot(epochs, val_loss, label='validation loss', c='green')
    axs[0].plot(epochs, test_loss, label='Test loss', c='blue')
    axs[0].set_title('Loss')
    axs[0].set_ylabel('Loss')
    axs[0].legend()

    # Accuracy plot
    axs[1].plot(epochs, train_acc, label='Training accuracy', c='red')
    axs[1].plot(epochs, val_acc, label='validation accuracy', c='green')
    axs[1].plot(epochs, test_acc, label='Test accuracy', c='blue')
    axs[1].set_title('Accuracy')
    axs[1].set_ylabel('Accuracy')
    axs[1].legend()

    plt.subplots_adjust(hspace=0.5)
    plt.show()
```

```python
plot_loss_acc(history_1)
```

```python
number_of_layers = 3
history_2 = {'train_loss':[], 'train_acc':[], 'val_loss':[], 'val_acc':[], 'test_loss':[], 'test_acc':[]}

while number_of_layers <= 20:
    model = keras.Sequential()
    model.add(layers.Input(shape=(28, 28)))
    model.add(layers.Flatten())
    for _ in range(number_of_layers - 1):
        model.add(layers.Dense(20, activation='relu'))
    model.add(layers.Dense(10, activation='softmax'))
    model.compile(loss=keras.losses.categorical_crossentropy,
              optimizer=keras.optimizers.Adam(learning_rate=6e-5),
              metrics=['accuracy'])
    history = model.fit(X_train, y_train, epochs=100, batch_size=32, validation_split=0.1)
    test_loss, test_acc = model.evaluate(X_test, y_test, batch_size=128)
    history_2['train_loss'].append(history.history['loss'][-1])
    history_2['train_acc'].append(history.history['accuracy'][-1])
    history_2['val_loss'].append(history.history['val_loss'][-1])
    history_2['val_acc'].append(history.history['val_accuracy'][-1])
    history_2['test_loss'].append(test_loss)
    history_2['test_acc']  .append(test_acc)
    number_of_layers += 1

```

```python
plot_loss_acc(history_2)
```
